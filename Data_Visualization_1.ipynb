```python
import matplotlib.pyplot as plt
students=["A","B","C","D","E"]
marks=[85,70,90,60,20]
```

```python
plt.bar(students,marks,color="skyblue")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Performance")
plt.show()
```

```python
hours=[2,3,4,5,6,7]
scores=[50,55,60,55,45,76]
plt.scatter(hours,scores,color="green")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Hours vs Scores")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,color="red")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
tips = pd.read_csv("/content/payment_fraud.csv")
sns.boxplot(x="accountAgeDays",y="paymentMethod",data=tips)
plt.title("Box PLot")
plt.show()
```

```python
import plotly.express as px
fig=px.scatter(tips,x="accountAgeDays",y="paymentMethod",color="paymentMethod",size="numItems",hover_data=["accountAgeDays", "paymentMethod", "numItems"])
fig.show()
```

```python
data={'Month':["Jan","Feb","Mar","Apr"],
      'Sales':[200,250,300,400]}
df=pd.DataFrame(data)
df.plot(x="Month",y="Sales",kind="line",marker="*")
plt.title("Monthly Sales Trend")
plt.show()
```
